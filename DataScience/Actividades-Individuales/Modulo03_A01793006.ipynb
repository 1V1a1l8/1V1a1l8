**AGRUPAMIENTO**

El método "groupby" agrupa los datos por diferentes categorías. Los datos se agrupan en función de una o varias variables, y el análisis se realiza en los grupos individuales.

```python
#Tomamos los valores unicos de la variable drive-wheels
df['drive-wheels'].unique()
```

```python
#Creamos un grupo en base a las clases unicas
df_group_one = df[['drive-wheels','body-style','price']]
```

```python
#Calculo de promedio de diferentes categorias
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
#Test agrupando por drive-wheels y body-style
df_gptest = df[['drive-wheels','body-style','price']] #Definimos las variables a probar
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean() #Calculamos la media de los precios
grouped_test1
```

```python
#Hacemos un pivot de la tabla para tener una lectura mas facil
#Drive-wheels  como registros y body-style se convierte en columnas
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
#Uso de heatmap para visualizar la relacion entre Body-style y Price
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#Las etiquetas predeterminadas no nos transmiten información útil, por lo que se modican
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

**Correlacion y causalidad**

Correlacion: una medida del grado de interdependencia entre variables.

Causalidad: la relación entre causa y efecto entre dos variables.


*Correlacion de Pearson*

Mide la dependencia lineal entre dos variables X e Y.
El coeficiente resultante es un valor entre -1 y 1 inclusive, donde:

  1: Correlacion lineal positiva perfecta.

  0: Sin correlación lineal, lo más probable es que las dos variables no se afecten entre sí.

  -1: Correlacion lineal negativa perfecta.

```python
#La relacion de Person se define con la funcion corr()
df.corr()
```

*Valor p*

Probabilidad de que la correlación entre dos variables sea estadísticamente significativa. Normalmente, elegimos un nivel de significancia de 0.05, lo que significa que estamos 95% seguros de que la correlación entre las variables es significativa.

Por convención, cuando el

* p < 0,001: decimos que hay pruebas sólidas de que la correlación es significativa.

* p < 0,05: existe evidencia moderada de que la correlación es significativa.

* p < 0,1: hay evidencia débil de que la correlación es significativa.

* p > 0,1: no hay evidencia de que la correlación sea significativa.

```python
#La funcion se llama stats y es de la libreria scripy

#Importamos libbrerias y llamamos a la funcion
from scipy import stats
```

**Ejercicios de Correlacion y Valor P**

Calculemos el coeficiente de correlación de Pearson y el valor P de la 'wheel-base' y el 'price'.

```python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

Dado que el valor p es < 0,001, la correlación entre la distancia entre ejes y el precio es estadísticamente significativa, aunque la relación lineal no es extremadamente fuerte (~0,585).

Calculemos el coeficiente de correlación de Pearson y el valor P de la 'horsepower' y el 'price'.

```python
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)  
```

Dado que el valor p es > 0,001, la correlación entre la potencia y el precio es estadísticamente no es significativa y la relación lineal es bastante fuerte (~0,809, cerca de 1).

**ANOVA**

Metodo estadístico utilizado para probar si existen diferencias significativas entre las medias de dos o más grupos. ANOVA devuelve dos parámetros:

F-test score: asume que las medias de todos los grupos son iguales, calcula cuánto se desvían las medias reales de la suposición y lo informa como la puntuación de la prueba F. Una puntuación mayor significa que hay una diferencia mayor entre las medias.

P-value: indica cuán estadísticamente significativo es nuestro valor de puntaje calculado.

```python
#Para ver si los diferentes tipos de 'ruedas motrices' impactan en el 'precio', agrupamos los datos.
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

Podemos usar la función 'f_oneway' en el modulo 'stats' para obtener el puntaje de la prueba F y el valor P.

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```
