#DATA WRANGLING

*Preprocesamiento:* Proceso de convertir o mapear datos de un formato "en bruto" a otro
formato que esté listo para su posterior análisis.

*Valor Faltante:* Una condición de "valor faltante" ocurre cada vez que una entrada de datos se deja vacía.

*Formato de Datos:* Estandariza los valores en el mismo formato, unidad o convención.

*Normalización:* Es una forma de llevar todos los datos a un rango similar, para una comparación más útil.

*Binning / Agrupamiento:* El agrupamiento crea categorías más grandes a partir de un conjunto de valores numéricos. Es particularmente útil para la comparación entre grupos de datos.

```python
#Importanción de Librerias
import pandas as pd
import matplotlib.pylab as plt
import requests
```

```python
#Almacenar archivo de forma local

filename=("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data")

def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        print(response.content)
        with open(filename, "wb") as f:
            f.write(response.content)

Downloads = download(filename, "imports-85.data") 
```

```python
#Asignación de Encabezados

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head() #Impresion de datos
```

*Sustitur valores Nulos: Se utiliza la función de replace*

```python
import numpy as np

df.replace("?", np.nan, inplace = True)
df.head(5)
```

*Revision de datos nulos, utilizando la funcion isnull()*

```python
missing_data = df.isnull()
missing_data.head(5)
```

*Utilizamos un ciclo para encontrar los valores perdidos en cada columna*

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

*Calculamos la media para normalizar las perdidas de datos en las columnas*

```python
#Promedio general
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Promedio de Normalización de Perdidas:", avg_norm_loss)
```

*Remplazar los vvaores "NaN" con el valor medio en la columna "pérdidas normalizadas"*

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

*Repetimos el proceso para la columna 'bore'*

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Promedio de bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

*Repetimos la normalizacion en la columna stroke*

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Promedio de stroke:", avg_stroke)

# Remplazamos los valores NaN por la media en la columna "stroke"
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

**Cambiar el formato de datos**

.dtype() para comprobar el tipo de datos

.astype() para cambiar el tipo de datos

```python
#Validar el tipo de datos 
df.dtypes
```

```python
#Convertimos el tipo de datos a los correctos
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#Validacion del cambio 
df.dtypes
```

**Estandarizacón de Datos**

Proceso de transformar los datos en un formato común, lo que permite al investigador hacer una comparación significativa.

```python
#Estandarizacion de la columna city-mpg
df['city-L/100km'] = 235/df["city-mpg"]

# Validación
df.head()
```

**Normalización de Datos**

La normalización es el proceso de transformar valores de varias variables en un rango similar.

```python
#Normalizacion de datos
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 

df[["length","width","height"]].head() #Mostrar datos
```

**Binning o Agrupamiento**

Proceso de transformación de variables numéricas continuas en 'contenedores' categóricos discretos para el análisis agrupado.

```python
#Generacion de los grupos linspace(start_value, end_value, numbers_generated
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#Asignacion de los nombres 
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts() #Conteo de los datos

#Creación de grafico
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
