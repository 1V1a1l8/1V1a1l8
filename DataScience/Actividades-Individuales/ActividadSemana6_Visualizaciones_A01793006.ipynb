#Reduccion de dimensiones con PCA

*Debido a que tenemos 12 variables numericas todas en la mmisma escala, consideramos importante hacer una reduccion de dimensiones para que el analisis posterior sea mas sencillo, ademas de conocer cuales son las variables mas significativas del conjunto.*

```python
#Importacion de Librerias
from sklearn.decomposition import PCA
#Se define el número de PCA a generar 
n=scaled_df.shape[1] #Se define en base a la cantidad de columnas del DF escalado

#encontrando los PCA
pca = PCA(n_components=n, random_state=1)
data_pca1 = pd.DataFrame(pca.fit_transform(scaled_df))

#varianza explicada por cada PCA
exp_var = pca.explained_variance_ratio_


# número mínimo de componentes que explican el 70% de la varianza
sum = 0
for ix, i in enumerate(exp_var):
  sum = sum + i
  if(sum>0.70):
    print("Número de PCA que explican al menos el 70% de la varianza: ", ix+1)
    break
  
```

*La cantidad de componentes que se necesitan para conservar el 70% de la varianza es de 4,sin embargo, decidi calcular la cantidad de variables necesarias para conservar el 90% de la varianza y disminuir la perdida.*

```python
# número mínimo de componentes que explican el 90% de la varianza
sum = 0
for ix, i in enumerate(exp_var):
  sum = sum + i
  if(sum>0.90):
    print("Número de PCA que explican al menos el 90% de la varianza: ", ix+1)
    break
  
```
