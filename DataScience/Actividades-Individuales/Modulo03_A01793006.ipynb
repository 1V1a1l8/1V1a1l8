**AGRUPAMIENTO**

El mtodo "groupby" agrupa los datos por diferentes categoras. Los datos se agrupan en funcin de una o varias variables, y el anlisis se realiza en los grupos individuales.

```python
#Tomamos los valores unicos de la variable drive-wheels
df['drive-wheels'].unique()
```

```python
#Calculo de promedio de diferentes categorias
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
#Test agrupando por drive-wheels y body-style
df_gptest = df[['drive-wheels','body-style','price']] #Definimos las variables a probar
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean() #Calculamos la media de los precios
grouped_test1
```

```python
#Hacemos un pivot de la tabla para tener una lectura mas facil
#Drive-wheels  como registros y body-style se convierte en columnas
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
#Uso de heatmap para visualizar la relacion entre Body-style y Price
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#Las etiquetas predeterminadas no nos transmiten informacin til, por lo que se modican
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

**Correlacion y causalidad**

Correlacion: una medida del grado de interdependencia entre variables.

Causalidad: la relacin entre causa y efecto entre dos variables.


*Correlacion de Pearson*

Mide la dependencia lineal entre dos variables X e Y.
El coeficiente resultante es un valor entre -1 y 1 inclusive, donde:

  1: Correlacion lineal positiva perfecta.

  0: Sin correlacin lineal, lo ms probable es que las dos variables no se afecten entre s.

  -1: Correlacion lineal negativa perfecta.

```python
#La relacion de Person se define con la funcion corr()
df.corr()
```

*Valor p*

Probabilidad de que la correlacin entre dos variables sea estadsticamente significativa. Normalmente, elegimos un nivel de significancia de 0.05, lo que significa que estamos 95% seguros de que la correlacin entre las variables es significativa.

Por convencin, cuando el

* p < 0,001: decimos que hay pruebas slidas de que la correlacin es significativa.

* p < 0,05: existe evidencia moderada de que la correlacin es significativa.

* p < 0,1: hay evidencia dbil de que la correlacin es significativa.

* p > 0,1: no hay evidencia de que la correlacin sea significativa.

```python
#La funcion se llama stats y es de la libreria scripy

#Importamos libbrerias y llamamos a la funcion
from scipy import stats
```

**Ejercicios de Correlacion y Valor P**

Calculemos el coeficiente de correlacin de Pearson y el valor P de la 'wheel-base' y el 'price'.

```python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

Dado que el valor p es < 0,001, la correlacin entre la distancia entre ejes y el precio es estadsticamente significativa, aunque la relacin lineal no es extremadamente fuerte (~0,585).

Calculemos el coeficiente de correlacin de Pearson y el valor P de la 'horsepower' y el 'price'.

```python
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)  
```

Dado que el valor p es > 0,001, la correlacin entre la potencia y el precio es estadsticamente no es significativa y la relacin lineal es bastante fuerte (~0,809, cerca de 1).

**ANOVA**

Metodo estadstico utilizado para probar si existen diferencias significativas entre las medias de dos o ms grupos. ANOVA devuelve dos parmetros:

F-test score: asume que las medias de todos los grupos son iguales, calcula cunto se desvan las medias reales de la suposicin y lo informa como la puntuacin de la prueba F. Una puntuacin mayor significa que hay una diferencia mayor entre las medias.

P-value: indica cun estadsticamente significativo es nuestro valor de puntaje calculado.
